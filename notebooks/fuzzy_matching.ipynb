```python
import pandas as pd
import difflib
from fuzzywuzzy import fuzz
```

```python
csv_file = 'rw.csv'
csv_dtype = {'reason': 'str'}
```

```python
# nrows = 1 * 1000
# cols = ['review_id','text','status','reason']
# df = pd.read_csv(csv_file, delimiter=",", dtype=csv_dtype, quotechar='\'', escapechar="\\", skiprows=0, nrows=nrows, names=cols)
# # df[['reason']] = df[['reason']].fillna('-')
# df = df[['text']]
# df = df.dropna(subset=['text'])
# display(df.head(10))

nrows = 1 * 5000000

df = pd.read_csv(csv_file, delimiter=",", nrows=nrows, dtype=csv_dtype, quotechar='\'', escapechar="\\", skiprows=1, names=['review_id','text','status','reason'])
df = df.dropna(subset=['text'])
df['reason'] = df[['reason']].fillna('-')

# ne_reason = ['contains non-english content', 'regional language review']
ne_reason = ['profane']
def is_ne(reason):
   for res in ne_reason:
       if res in reason: return True
   return False

df['ne_label'] = df.apply(lambda row: 1 if is_ne(row['reason'].lower()) else 0, axis=1)
df = df[df['ne_label'] == 1]
```

```python
df = df[['text']]
df.head(10)
print(df.size)
```

```python
hindi_profane_dict = '../datasets/hindi_profane_words.csv'
```

```python
profane_words_o = (line.strip() for line in open(hindi_profane_dict))

profane_words = []
for w in profane_words_o:
    profane_words.append(w)
```

```python
reviews = list(df['text'])
```

```python
threshold = 50
count = 0
for review_text in reviews:
    for profane_word in profane_words:
        match = fuzz.token_sort_ratio(review_text, profane_word)
        if match > threshold:
            count += 1
            print("{}, {}, {}".format(match, profane_word, review_text))
            break
```

```python
print("{} profane reviews in total of {} reviews".format(count, len(reviews)))
```
